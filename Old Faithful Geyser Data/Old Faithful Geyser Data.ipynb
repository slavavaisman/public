# Old Faithful Geyser Data data analysis

## A famous dataset on geyser explosions (also available as a built-in dataset for R. HÃ¤rdle, W. (1991), Smoothing Techniques with Implementation in S. New York: Springer.). Each row depicts an observed eruption of the Old Faithful Geyser in Yellowstone National Park. The eruptions column shows the duration of the eruption in minutes, while the waiting column reflects the time before the next eruption. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# load data
df = pd.read_csv("faithful.csv")
df = df[["better.eruptions","waiting"]]
df.columns = ["erupt","waiting"]
df_train, df_test = train_test_split(df, test_size=0.33, random_state=42)
```

```python
#print(df_train.head())
print("train size - ",len(df_train))
print("test size - ",len(df_test))
sns.scatterplot(data=df_train, x="erupt", y="waiting")
```

```python
# create train/test arrays
X_train = df_train.erupt.values.reshape(-1,1)
y_train = df_train.waiting

X_test = df_test.erupt.values.reshape(-1,1)
y_test = df_test.waiting
```

# Constant model



```python
# Constant model
y_hat = np.mean(y_train)*np.ones(len(y_test))
const_mse = mean_squared_error(y_test, y_hat)
print("Constant model test mean_squared_error : ",const_mse)
```

# Linear regression model

```python
# Linear regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
y_hat = reg.predict(X_test)
print("coeficients : ",reg.intercept_, reg.coef_)
linreg_mse = mean_squared_error(y_test, y_hat)
print("Linear Regression test mean_squared_error : ",linreg_mse)
sns.regplot(data=df, x="erupt", y="waiting");
```

# Random Forest Regressor model

```python
# Random Forest Regressor model
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor().fit(X_train, y_train)
y_hat = reg.predict(X_test)
rf_mse = mean_squared_error(y_test, y_hat)
print("Random Forest Regressor test mean_squared_error : ",rf_mse)
```

```python
# if we look at the output variable, it looks like a mixture of Gaussians
sns.histplot(data=df_train, x = df_train.waiting)
```

## It seems like the regression can benefit from a mixture distribution and this is indeed the case as shown below.

# Mixture of Gaussians regression

```python
from gmr import GMM
bX_train = np.concatenate((X_train.flatten().reshape(-1,1), y_train.values.reshape(-1,1)), axis = 1)
#gmm = GMM(n_components=2, random_state=0)
gmm = GMM(n_components=2)
gmm.from_samples(bX_train)
X_test.shape
y_hat = gmm.predict([0],X_test)
gmm_mse = mean_squared_error(y_test, y_hat)

print("Mixture of Gaussians regression test mean_squared_error : ",gmm_mse)
```

```python
print("Constant model test mean_squared_error : ",const_mse)
print("Linear Regression test mean_squared_error : ",linreg_mse)
print("Random Forest Regressor test mean_squared_error : ",rf_mse)
print("Mixture of Gaussians regression test mean_squared_error : ",gmm_mse)
```
